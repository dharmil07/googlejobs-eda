# Exploratory Data Analysis of what kind of Jobs and Job Locations are provided by Google

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
google = pd.read_csv('C:\Dharmil Imp\Projects\Google Job\job_skills.csv')

google.shape
```

```python
# Companies involved

plt.rcParams['figure.figsize'] = (7, 9)
google['Company'].value_counts().plot.bar(color = 'blue')
plt.title('The companies Involved for recruitment', fontsize = 20)
plt.xlabel('Names of companies', fontsize = 15)
plt.show()
```

```python
# Most Popular 20 Job Titles of Google

plt.rcParams['figure.figsize'] = (19, 10)
google['Title'].value_counts().sort_values(ascending = False).head(20).plot.bar(color = 'pink')
plt.title("Most Popular 20 Job Titles of Google", fontsize = 20)
plt.xlabel('Names of Job Titles', fontsize = 15)
plt.ylabel('count', fontsize = 15)
plt.show()
```

```python
#Most Popular 20 Job Categories of Google

plt.rcParams['figure.figsize'] = (15, 6)
google['Category'].value_counts().sort_values(ascending = False).head(20).plot.bar(color = 'lightgreen')
plt.title("Most Popular 20 Job Categories of Google", fontsize = 20)
plt.xlabel('Names of Job Categories', fontsize = 15)
plt.ylabel('count', fontsize = 15)
plt.show()
```

```python
# Most Popular 20 Job Locations of Google

plt.rcParams['figure.figsize'] = (17, 6)
google['Location'].value_counts().sort_values(ascending = False).head(20).plot.bar(color = 'lightblue')
plt.title("Most Popular 20 Job Locations of Google", fontsize = 20)
plt.xlabel('Names of Job Locations', fontsize = 15)
plt.ylabel('count', fontsize = 15)
plt.show()
```

```python
# Places with Most Jobs

from wordcloud import WordCloud
from wordcloud import STOPWORDS

stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'white').generate(str(google['Location']))

print(wordcloud)
plt.rcParams['figure.figsize'] = (12, 12)
plt.axis('off')
plt.title('Places with Most Jobs', fontsize = 30)
plt.imshow(wordcloud)
plt.show()
```

```python
# Most Popular Responsibilities

from wordcloud import WordCloud
from wordcloud import STOPWORDS

stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'pink', max_words = 100).generate(str(google['Responsibilities']))

print(wordcloud)
plt.rcParams['figure.figsize'] = (15, 15)
plt.axis('off')
plt.imshow(wordcloud)
plt.title('Most Popular Responsibilities', fontsize = 30)
plt.show()
```

```python
# Most Popular Educational Requirements

from wordcloud import WordCloud
from wordcloud import STOPWORDS

stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'cyan', max_words = 80).generate(str(google['Minimum Qualifications']))

print(wordcloud)
plt.rcParams['figure.figsize'] = (12, 12)
plt.axis('off')
plt.imshow(wordcloud)
plt.title('Most Popular Educational Requirements', fontsize = 30)
plt.show()
```
